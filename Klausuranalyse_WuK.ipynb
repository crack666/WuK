## 🧮 **Formelsammlung für DIN A4 Blatt**

### **KOMBINATORIK**
```
Permutationen mit Wiederholung: n!/(n₁!·n₂!·...·nₖ!)
Kombinationen: C(n,k) = n!/(k!(n-k)!)
```

### **WAHRSCHEINLICHKEITSRECHNUNG**
```
Binomialverteilung: P(X=k) = C(n,k)·p^k·(1-p)^(n-k)
Bayes: P(A|B) = P(B|A)·P(A)/P(B)
Erwartungswert: E[X] = Σ x·P(X=x)
Varianz: Var(X) = E[X²] - (E[X])²
```

### **MODULARE ARITHMETIK**
```
Euler'sches Theorem: a^φ(n) ≡ 1 (mod n), falls gcd(a,n)=1
φ(n) = n·∏(1-1/p) für alle Primteiler p von n
φ(p^k) = p^(k-1)·(p-1)
Erweiterte Euklidische: ax + by = gcd(a,b)
```

### **RSA-SPEZIALFÄLLE**
```
Standard: φ(n) = (p-1)(q-1)
Quadratzahl: φ(p²) = p(p-1)
Öffentlicher Exponent: gcd(e,φ(n)) = 1
Private Schlüssel: d·e ≡ 1 (mod φ(n))
Entschlüsselung: m = c^d mod n
```

### **AFFINE CHIFFRE**
```
f(x) = (ax + b) mod 26
Gleichungssystem: f(x₁) = y₁, f(x₂) = y₂
Lösung: a·(x₁-x₂) ≡ y₁-y₂ (mod 26)
```

### **STETIGE VERTEILUNGEN**
```
Normierung: ∫f(x)dx = 1
E[X] = ∫x·f(x)dx
Var(X) = ∫x²·f(x)dx - (E[X])²
```

---

## 🎯 **Kritische Spezialfälle für RSA**

**WICHTIG:** Bei φ(n) = 25 ist n = 5² = 25, daher φ(25) = 5·(5-1) = 20, **NICHT** (5-1)·(5-1) = 16!

**Allgemeine Regel für φ(n):**
- n = p·q (verschiedene Primzahlen): φ(n) = (p-1)(q-1)
- n = p² (Primzahlquadrat): φ(n) = p(p-1)
- n = p^k: φ(n) = p^(k-1)(p-1)

## 🧮 Mathematische Formeln für DIN A4 Formelsammlung

### 1. KOMBINATORIK - Buchstaben/Ziffern-Anordnungen

**Hauptformel:**
$$\text{Permutationen mit Wiederholung} = \frac{n!}{n_1! \cdot n_2! \cdot \ldots \cdot n_k!}$$

**Beispiel:** Für A,T,T,T,E,E (6 Buchstaben: 1×A, 3×T, 2×E)
$$\frac{6!}{1! \cdot 3! \cdot 2!} = \frac{720}{1 \cdot 6 \cdot 2} = 60$$

**Spezialfälle:**
- Beginnt mit X und endet mit Y: $\frac{(n-2)!}{n_1! \cdot n_2! \cdot \ldots}$
- Gleiche Elemente zusammen: Behandle als ein Element

---

### 2. AUSWAHL-KOMBINATORIK

**Hauptformel:**
$$\binom{n}{k} = C(n,k) = \frac{n!}{k!(n-k)!}$$

**Spezialfälle:**
- Pflichtauswahl: $\binom{n-p}{k-p}$ (p Pflichtelemente)
- Mindestens m aus ersten r: $\sum_{i=m}^{\min(r,k)} \binom{r}{i} \cdot \binom{n-r}{k-i}$

---

### 3. FAHRSTUHL-AUFGABEN

**Grundprinzip:** Gleichverteilung auf Stockwerke

**Formeln:**
- $n$ Personen, $k$ Stockwerke
- Gesamtmöglichkeiten: $k^n$
- Nur ein Halt: $k \cdot \binom{n}{1} = k \cdot n$ (alle in einem Stockwerk)
- Alle verschiedene Stockwerke: $k \cdot (k-1) \cdot \ldots \cdot (k-n+1)$

---

### 4. WAHRSCHEINLICHKEITSRECHNUNG

**Binomialverteilung:**
$$P(X = k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k}$$

**Bayes'sche Formel:**
$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$

**Erwartungswert und Varianz:**
- $E[X] = \sum_{i} x_i \cdot P(X = x_i)$
- $\text{Var}(X) = E[X^2] - (E[X])^2$
- $\sigma = \sqrt{\text{Var}(X)}$

**Stetige Verteilungen:**
- Normierung: $\int_{-\infty}^{\infty} f(x) dx = 1$
- $E[X] = \int_{-\infty}^{\infty} x \cdot f(x) dx$
- $\text{Var}(X) = \int_{-\infty}^{\infty} x^2 \cdot f(x) dx - (E[X])^2$

---

### 5. MODULARE ARITHMETIK

**Euler'sches Theorem:**
$$a^{\varphi(n)} \equiv 1 \pmod{n} \quad \text{falls } \gcd(a,n) = 1$$

**Euler'sche Funktion:**
$$\varphi(n) = n \prod_{p|n} \left(1 - \frac{1}{p}\right)$$

**Spezialfälle:**
- $\varphi(p) = p-1$ (p prim)
- $\varphi(p^k) = p^{k-1}(p-1)$
- $\varphi(p \cdot q) = (p-1)(q-1)$ (p,q prim, p≠q)

**Erweiterter Euklidischer Algorithmus:**
$$ax + by = \gcd(a,b)$$

---

### 6. RSA-KRYPTOGRAPHIE ⚠️ **KRITISCHE SPEZIALFÄLLE**

**Standard-RSA:**
- $n = p \cdot q$ (verschiedene Primzahlen)
- $\varphi(n) = (p-1)(q-1)$
- $e \cdot d \equiv 1 \pmod{\varphi(n)}$
- Entschlüsselung: $m = c^d \bmod n$

**🚨 WICHTIGER SPEZIALFALL:**
- $n = p^2$ (Primzahlquadrat): $\varphi(n) = p(p-1)$
- **Beispiel:** $n = 25 = 5^2 \Rightarrow \varphi(25) = 5 \cdot 4 = 20$
- **NICHT** $(5-1)(5-1) = 16$!

**RSA-Algorithmus:**
1. $n$ faktorisieren: $n = p^a \cdot q^b \cdot \ldots$
2. $\varphi(n)$ berechnen
3. $\gcd(e, \varphi(n)) = 1$ prüfen
4. $d$ mit $e \cdot d \equiv 1 \pmod{\varphi(n)}$ finden
5. $m = c^d \bmod n$ berechnen

---

### 7. AFFINE CHIFFRE

**Verschlüsselungsfunktion:**
$$f(x) = (ax + b) \bmod 26$$

**Lösungsweg:**
1. Zwei Buchstaben-Paare gegeben: $(x_1, y_1)$ und $(x_2, y_2)$
2. Gleichungssystem aufstellen:
   - $ax_1 + b \equiv y_1 \pmod{26}$
   - $ax_2 + b \equiv y_2 \pmod{26}$
3. Subtrahieren: $a(x_1 - x_2) \equiv y_1 - y_2 \pmod{26}$
4. $a$ lösen (modulare Inverse)
5. $b$ aus einer Gleichung bestimmen

**Wichtig:** $\gcd(a, 26) = 1$ muss gelten!

---

### 8. HÄUFIGE ZAHLENWERTE

**Modulare Inverse mod 26:**
- $1^{-1} \equiv 1$, $3^{-1} \equiv 9$, $5^{-1} \equiv 21$
- $7^{-1} \equiv 15$, $9^{-1} \equiv 3$, $11^{-1} \equiv 19$
- $15^{-1} \equiv 7$, $17^{-1} \equiv 23$, $19^{-1} \equiv 11$
- $21^{-1} \equiv 5$, $23^{-1} \equiv 17$, $25^{-1} \equiv 25$

**Buchstaben-Zuordnung:**
A=0, B=1, C=2, D=3, E=4, F=5, G=6, H=7, I=8, J=9, K=10, L=11, M=12, N=13, O=14, P=15, Q=16, R=17, S=18, T=19, U=20, V=21, W=22, X=23, Y=24, Z=25

## 🎯 Lösungsstrategien & Checkliste

### RSA-Entschlüsselung Schritt-für-Schritt:

1. **Faktorisierung prüfen:**
   - $n = p \cdot q$ → $\varphi(n) = (p-1)(q-1)$
   - $n = p^2$ → $\varphi(n) = p(p-1)$ ⚠️
   - $n = p^k$ → $\varphi(n) = p^{k-1}(p-1)$

2. **Öffentlichen Exponenten bestimmen:**
   - $\gcd(e, \varphi(n)) = 1$ prüfen
   - Gegeben: mehrere Möglichkeiten → alle testen

3. **Privaten Schlüssel finden:**
   - Erweiterte Euklidische Algorithmus: $e \cdot d \equiv 1 \pmod{\varphi(n)}$

4. **Entschlüsseln:**
   - $m = c^d \bmod n$

### Bayes'sche Aufgaben:

**Schema:**
- Gegeben: $P(\text{Krankheit}) = p$, $P(\text{Test+}|\text{Krank}) = s$, $P(\text{Test+}|\text{Gesund}) = f$
- Gesucht: $P(\text{Krank}|\text{Test+})$

**Lösung:**
$$P(\text{Krank}|\text{Test+}) = \frac{s \cdot p}{s \cdot p + f \cdot (1-p)}$$

#### **Bayes-Schema (immer gleich):**
1. **Gegeben:** Grundwahrscheinlichkeit P(Krankheit), Sensitivität, Falsch-Positiv-Rate
2. **Gesucht:** P(Krankheit | Test positiv)
3. **Formel:** Immer dieselbe Bayes-Formel
4. **Keine Sonderfälle** wie bei RSA (φ-Funktion) oder Affine Chiffre

#### **Bayes in allen Klausuren:**
- **21SoSe:** Defekte Produkte (3% defekt, 90% erkannt, 2% falsch aussortiert)
- **SoSe20:** Viruserkrankung (10⁻⁵ Wahrscheinlichkeit, 99% erkannt, 0,5% falsch-positiv)  
- **SS14:** Seltene Krankheit (10⁻⁵ Wahrscheinlichkeit, 95% erkannt, 0,5% falsch-positiv)

**Immer dasselbe Schema!** Nur andere Zahlen einsetzen.

#### **Bayes-Erfolgs-Formel:**
$$P(\text{Krankheit}|\text{Test+}) = \frac{P(\text{Test+}|\text{Krank}) \cdot P(\text{Krank})}{P(\text{Test+}|\text{Krank}) \cdot P(\text{Krank}) + P(\text{Test+}|\text{Gesund}) \cdot P(\text{Gesund})}$$

**Einmal verstanden = immer lösbar!**

### 🧠 **BAYES-ERFOLGS-REZEPT** (8-10 Punkte sicher)

#### **Schritt-für-Schritt Anleitung:**

1. **Gegeben identifizieren:**
   - P(Krankheit) = ? (meist sehr klein: 10⁻⁵, 0,03, etc.)
   - P(Test+ | Krank) = ? (Sensitivität: 90%, 95%, 99%)
   - P(Test+ | Gesund) = ? (Falsch-Positiv: 0,5%, 2%, etc.)

2. **Schema anwenden:**
   ```
   P(Krank | Test+) = (Sensitivität × P(Krank)) / 
                      (Sensitivität × P(Krank) + Falsch-Positiv × P(Gesund))
   ```

3. **Zahlenwerte einsetzen:**
   - P(Gesund) = 1 - P(Krank)
   - Alles in Dezimalbrüche umwandeln
   - Sorgfältig rechnen

4. **Antwort interpretieren:**
   - Meist sehr kleine Wahrscheinlichkeit trotz positivem Test!

#### **Typische Bayes-Zahlen:**
- **P(Krankheit):** 10⁻⁵ = 0,00001 oder 3% = 0,03
- **Sensitivität:** 90% = 0,9 oder 95% = 0,95
- **Falsch-Positiv:** 0,5% = 0,005 oder 2% = 0,02
